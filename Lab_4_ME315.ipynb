```r
library(tidyverse)
library(dplyr)
library(readxl)
```

```r
fname <- "/content/WDIEXCEL.xlsx"
dado <- read_excel(fname)
file.exists(fname)
fname
```

```r
countries <-  read_excel(fname, sheet = "Country")
```

```r
paises <- countries %>% filter(`Short Name` %in% c("Aruba", "Andorra", "Brazil", "Cuba", "Fiji"))
```

```r
WDI <-  read_excel(fname, sheet = "Data")
```

```r
WDI_3 <- WDI %>% select(`Country Code`, `Indicator Code`, `1960`:`2018` )
```

```r
WDI_tidy <- pivot_longer(WDI, cols = `1960`:`2018`, names_to = "Year") %>% drop_na()  %>% pivot_wider(c(`Country Code`, Year), names_from = "Indicator Code", values_from = "value") %>% mutate(Year = as.integer(Year))
WDI_tidy
```

```r
class(WDI_tidy$Year)
```

```r
CC <-countries %>% select(`Country Code`, `Short Name`) %>% left_join(WDI_tidy %>%  select(`Country Code`, Year , SP.DYN.CBRT.IN), by = "Country Code")
```

```r
CC <- CC %>% arrange(Year)
```

```r
ggplot(CC) + geom_line(aes(x = Year, y = SP.DYN.CBRT.IN, col = `Country Code`))
```
