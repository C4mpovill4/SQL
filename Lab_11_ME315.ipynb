```python
import polars as pl
```

```python
diamonds = pl.read_csv("https://me315-unicamp.github.io/dados/diamonds.csv.gz")
```

```python
diamonds.head(5)
```

```python
from plotnine import *
```

```python
from plotnine import  ggplot, aes, geom_point, geom_smooth, labs, theme_minimal

plot = (
    ggplot(diamonds, aes(x='carat', y='price')) +
    geom_point(alpha=0.6, color='blue') +
    labs(title='Relação entre Peso e Preço',
         x='Peso',
         y='Preço (USD)') +
    theme_minimal()
)

print(plot)
```

```python
plot = (
    ggplot(diamonds, aes(x='carat', y='price', color='cut')) +
    geom_point(alpha=0.6) +
    labs(title='Relação entre Peso e Preço para diferentes cortes de Diamante',
         x='Peso (Carat)',
         y='Preço (USD)') +
    theme_minimal()
)

plot
```

```python
plot = (
    ggplot(diamonds, aes(x='carat', y='price', color='cut', shape='clarity')) +
    geom_point(alpha=0.6) + facet_wrap('color') +
    labs(title='Relação entre Peso e Preço para diferentes Configurações de Diamante', x='Peso)', y='Preço (USD)') +
    theme_minimal()
)

plot
```

```python
plot = (
    ggplot(diamonds, aes(x='carat', y='price', color='cut', shape='clarity'))
    + geom_point(alpha=0.4, size=0.5)
    + facet_grid('cut ~ color')
    + labs(
        title='Relação entre Peso e Preço para diferentes Configurações de Diamante',
        x='Peso',
        y='Preço (USD)',
    )
    + theme_minimal()
    + scale_color_brewer(type='qual', palette='Set1')
)

plot
```
