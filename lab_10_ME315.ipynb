```python
pl.read_database("""
SELECT originalTitle, averageRating, numVotes
FROM ratings
INNER JOIN basics ON ratings.tconst = basics.tconst
ORDER BY averageRating DESC, numVotes
LIMIT 5
""", conn)
```

```python
pl.read_database("""
SELECT genres, COUNT(*) AS n FROM ratings
INNER JOIN basics ON ratings.tconst = basics.tconst
WHERE averageRating > 8 AND titleType = 'movie'
GROUP BY genres
ORDER BY n DESC
LIMIT 1
""", conn)
```
