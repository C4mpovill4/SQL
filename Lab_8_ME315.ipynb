```python
!wget https://me315-unicamp.github.io/dados/renda_adulta.csv.gz

import polars as pl

colunas = [
    "age", "workclass", "fnlwgt", "education", "education-num", "marital-status",
    "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss",
    "hours-per-week", "native-country", "income"
]
df = pl.read_csv(
    'renda_adulta.csv.gz',
    has_header=False,
    new_columns=colunas,
    null_values=["?"]
)


renda2 = df
renda2

renda = pl.read_csv(
    'renda_adulta.csv.gz',
    has_header=False,
    null_values=["?"]
)

renda = renda.rename({"column_10":"sex","column_15":"income"})

df.dtypes

df.shape

renda2.group_by("income").len()

media_horas_por_classe = (renda2.group_by("income")
    .agg(
    pl.col("hours-per-week")
    .mean()
    .alias("media_horas")
))
media_horas_por_classe

(
    renda2.group_by("occupation")
    .agg(pl.count().alias("num_pessoas"))
    .sort("num_pessoas", descending=True)
)

#ou
(
    renda2.drop_nulls("occupation")
    .group_by("occupation")
    .len().rename({"len":"n"})
    .sort("n", descending=True)
)

from plotnine import ggplot, geom_bar, aes,labs, theme_minimal

(
    ggplot(media_horas_por_classe,aes(x="income", y = "media_horas", fill = "income"))+
    geom_bar(stat = "identity")
    + theme_minimal()
    +labs(x = "Salario Anual", y = "Média de horas por semana", fill = "Classe de renda")
)


media_salario_por_genero = (
    renda2.group_by("sex")
    .agg(
        pl.col("hours-per-week").mean().alias("media_horas")
    )
)

(
    ggplot(media_salario_por_genero, aes(x="sex", y="media_horas", fill="sex"))
    + geom_bar(stat="identity")
    + theme_minimal()
    + labs(x="Sexo", y="Média de horas por semana", fill="Sexo")
)


renda_longo = (
    renda2
    .melt(
        id_vars=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'hours-per-week', 'native-country', 'income'],
        value_vars=['capital-gain', 'capital-loss'],
        variable_name='tipo',
        value_name='Valor'
    )
)
renda_longo
```
